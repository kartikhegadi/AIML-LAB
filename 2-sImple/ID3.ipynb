```python
#Import
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
#Dataset
golf_data = pd.read_csv("data3.csv")
golf_data_num = golf_data.apply(LabelEncoder().fit_transform)
data = np.array(golf_data_num.iloc[:, :4])
target = np.array(golf_data_num.iloc[:, 4])
```

```python
#ID3
golf_tree = DecisionTreeClassifier(criterion='entropy')
model = golf_tree.fit(data, target)

fn=['Outlook','Temperature','Humidity','Windy']
cn=['Yes', 'No']
```

```python
#Visualization
fig = plt.figure(figsize=(20,20))
_ =tree.plot_tree(golf_tree,
feature_names = fn,
class_names=cn,
filled = True)
plt.suptitle("ID3 on Golf Play Dataset")
plt.show()
```
